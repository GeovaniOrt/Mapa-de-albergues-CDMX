```python
try:
  data = pd.read_csv('/content/datos de albergues.csv', encoding='latin-1')
except FileNotFoundError:
  print("Error: El archivo albergues_cdmx.csv no se encontrÃ³.")
  exit()
# 2. Crear el mapa centrado en la CDMX (puedes ajustar las coordenadas y el zoom)
mapa_cdmx = folium.Map(location=[19.4326, -99.1332], zoom_start=11)

# 3. Iterar a travÃ©s de cada fila del DataFrame para agregar los marcadores
for index, row in data.iterrows():
    latitud = row['latitud']
    longitud = row['longitud']
    nombre = row.get('nombrerefugio')  # Obtener el nombre si existe, sino usar 'Albergue'
    capacidad = row['capacidad']
    alcaldia = row['alcaldia']

    # Crear el texto emergente (popup) para cada marcador
    popup_text = f"Nombre:{nombre} Capacidad: {capacidad} personas  Alcaldia: {alcaldia}"
    popup = folium.Popup(popup_text, parse_html=True)

    #Colores
        # Determinar color basado en la capacidad
    if capacidad <= 100:
        circle_color = 'blue'
    elif capacidad <= 300:
        circle_color = 'green'
    elif capacidad <= 600:
        circle_color = 'yellow'
    elif 600 < capacidad <= 800:  # CondiciÃ³n corregida
        circle_color = 'orange'
    else:
        circle_color = 'red'

    # Crear el marcador y aÃ±adirlo al mapa
    folium.CircleMarker(
        location=[latitud, longitud],
        radius=capacidad / 20,  # Ajusta el factor para el tamaÃ±o del cÃ­rculo
        color=circle_color,
        fill=True,
        fill_color=circle_color,
        fill_opacity=0.7,
        popup=popup
    ).add_to(mapa_cdmx)

# 4. Mapa visual
cloudflared tuneel -- file:///C:/Users/geova/Downloads/Mapa_Web/mapa_albergues_cdmx.html.html--

print("Mapa de albergues generado exitosamente como mapa_albergues_cdmx.html")
```
